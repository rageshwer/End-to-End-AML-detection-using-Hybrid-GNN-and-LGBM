```python
import shap
import pandas as pd
import numpy as np
import lightgbm as lgb
```

```python
model=lgb.Booster(model_file='/Users/rageshwer/Goal ML/Projects/AML - GNN/req_data/model.txt')
X_test=pd.read_parquet('/Users/rageshwer/Goal ML/Projects/AML - GNN/req_data/test_vectors.parquet')
```

```python
tree_explainer=shap.TreeExplainer(model)
shap_obj=tree_explainer(X_test)
shap_val_matrix=shap_obj.values
```

```python
lookup_df=pd.DataFrame()
lookup_df['GNN Embeddings']=shap_val_matrix[:,0:256].sum(axis=1)
lookup_df['Local Features']=shap_val_matrix[:,256:350].sum(axis=1)
lookup_df['One Hop Features']=shap_val_matrix[:,350:421].sum(axis=1)
lookup_df['Base Value']=shap_obj.base_values
```

```python
lookup_df.to_parquet('/Users/rageshwer/Goal ML/Projects/AML - GNN/req_data/lookup.parquet')
```
